Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/export?format=csv&gid=1724592523"
users = pd.read_csv(url)

# Видалення усіх стовпців з пропусками, крім стовпця "Мова програмування"
cols_with_na = users.columns[users.isnull().any()]
cols_to_drop = [col for col in cols_with_na if col != "Мова програмування"]
users = users.drop(columns=cols_to_drop)

# Видалення усіх рядків у вихідній таблиці
data = users.dropna()

# Створення нової таблиці python_data
python_data = data[data["Мова програмування"] == "Python"]

# Групування за стовпчиком "Посада"
grouped = python_data.groupby("Посада").size()

# Виконання агрегації даних за і знаходження мінімального та максимального значення у стовпчику "Зарплата.в.місяць"
salary_stats = python_data.groupby("Посада").agg(
    min_salary=("Зарплата на місяць", "min"),
    max_salary=("Зарплата на місяць", "max")
)

print(salary_stats)
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/export?format=csv&gid=1724592523"
users = pd.read_csv(url)

# Видалення усіх стовпців з пропусками, крім стовпця "Мова програмування"
cols_with_na = users.columns[users.isnull().any()]
cols_to_drop = [col for col in cols_with_na if col != "Мова програмування"]
users = users.drop(columns=cols_to_drop)

# Видалення усіх рядків у вихідній таблиці
data = users.dropna()

# Створення нової таблиці python_data
python_data = data[data["Мова програмування"] == "Python"]

# Групування за стовпчиком "Посада"
grouped = python_data.groupby("Посада").size()

# Виконання агрегації даних за і знаходження мінімального та максимального значення у стовпчику "Зарплата.в.місяць"
salary_stats = python_data.groupby("Посада").agg(
    min_salary=("Зарплата на місяць", "min"),
    max_salary=("Зарплата на місяць", "max")
)

# Cереднє значення заробітної плати на місяць та створення нового стовпчика "avg"
def fill_avg_salary(row):
    avg_salary = python_data["Зарплата на місяць"].mean()
    return avg_salary

python_data["avg"] = python_data.apply(fill_avg_salary, axis=1)
print(python_data[["Зарплата на місяць", "avg"]].head())
```
