Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
import pandas as pd

users = pd.read_csv("bestsellers with categories.csv")
users.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

author_books = users.groupby('author')['user_rating'].agg('mean')

print(author_books)
```

```python
import pandas as pd

users = pd.read_csv("bestsellers with categories.csv")
users.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

author_books = users.groupby('author')['user_rating'].agg('mean')

print(author_books.idxmin())
print(author_books.min())
```

З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг 
(Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
import pandas as pd

users = pd.read_csv("bestsellers with categories.csv")
users.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

author_books_1 = users.groupby('author')['name'].agg('count')
author_books_2 = users.groupby('author')['user_rating'].agg('mean')

books = pd.concat([author_books_1, author_books_2], axis = 1)

print(books)
```

Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

```python
import pandas as pd

users = pd.read_csv("bestsellers with categories.csv")
users.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

author_books_1 = users.groupby('author')['name'].agg('count')
author_books_2 = users.groupby('author')['user_rating'].agg('mean')

books = pd.concat([author_books_1, author_books_2], axis = 1)
books = books.sort_values(by=['name', 'user_rating'])
print(books)
```

Який автор перший у списку? Відповідь: Muriel Barbery

ГРАФІКИ

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

users = pd.read_csv("bestsellers with categories.csv")
top_authors = users['Author'].value_counts().nlargest(20).index
users_top = users[users['Author'].isin(top_authors)]

plt.figure(figsize=(16,8))
sns.countplot(y="Author", data=users_top, color="mediumseagreen")
plt.xlabel("Кількість книг")
plt.ylabel("Автор")
plt.title("Топ 20 авторів по кількості книг")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
author_counts = users_top['Author'].value_counts()
plt.pie(author_counts, labels=author_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Reds", 20))
plt.title("Топ 20 авторів по кількості книг")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(y=author_counts.index, x=author_counts.values, color="purple")
plt.xlabel("Кількість книг")
plt.ylabel("Автор")
plt.title("Топ 20 авторів по кількості книг")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(x="Author", y="User Rating", data=users_top)
plt.xticks(rotation=90)
plt.xlabel("Автор")
plt.ylabel("Рейтинг книги")
plt.title("Розподіл рейтингу книг у топ-20 авторів")
plt.tight_layout()
plt.show()
```
