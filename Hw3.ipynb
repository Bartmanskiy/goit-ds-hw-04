```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv("Housing.csv")
X = data[['area', 'bedrooms', 'bathrooms']].values
y = data['price'].values

X_b = np.c_[np.ones((X.shape[0], 1)), X] #  Додаємо стовпець одиниць
```

Напишіть функцію гіпотези лінійної регресії у векторному вигляді

```python
def predict(X, w):
    return X @ w 
```

Створіть функцію для обчислення функції втрат у векторному вигляді

```python
def compute_loss(X, y, w): 
    m = len(y)
    predictions = predict(X, w)
    return (1/(2*m)) * np.sum((predictions - y)**2)
```

Реалізуйте один крок градієнтного спуску

```python
def gradient_step(X, y, w, lr=0.01):
    m = len(y)
    predictions = predict(X, w)
    gradient = (1/m) * X.T @ (predictions - y)
    w_new = w - lr * gradient
    return w_new
```

Знайдіть найкращі параметри w для датасету використовуючи написані вами функції, прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень

```python
w = np.zeros(X_b.shape[1])
lr = 1e-9
epochs = 100000

for i in range(epochs):
    w = gradient_step(X_b, y, w, lr)

print("Параметри w (градієнтний спуск):", w)
print("Фінальна функція втрат:", compute_loss(X_b, y, w))
```

Знайдіть ці ж параметри за допомогою аналітичного рішення

```python
w_analytical = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print("Параметри w (аналітичне рішення):", w_analytical)
```

Для перевірки спрогнозованих значень, використайте LinearRegression з бібліотеки scikit-learn та порівняйте результати

```python
model = LinearRegression()
model.fit(X, y)
print("Параметри w (sklearn):", np.r_[model.intercept_, model.coef_])
```
